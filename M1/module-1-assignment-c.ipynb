FASTA format is a file format used to exchange information between genetic sequence databases.  A sequence in FASTA format begins with a single-line description, followed by lines of sequence data.  The description line is distinguished from the sequence data by a greater-than ("`>`") symbol in the first column.   

The word following the "`>`" symbol is the identifier of the sequence, and the rest of the line is the description (both are optional). There should be no space between the ">" and the first letter of the identifier. It is recommended that all lines of text be shorter than 80 characters. No blank lines are allowed in the middle of the FASTA input.  The sequence ends if another line starting with a "`>`" appears; this indicates the start of another sequence. 

Normally, identifiers are simply protein accession, name or Entrez gi's (e.g., `Q5I7T1`, `AG10B_HUMAN`, `129295`), but a bar-separated NCBI sequence identifier (e.g., `gi|129295`) will also be accepted. Any arbitrary user-specified sequence identifier can also be used (e.g., CLONE00073452) but you are advised to use sufficiently long unique words in such case. There should be no space between the "`>`" and the first letter of the identifier. It is recommended that all lines of text be shorter than 80 characters in length.

- Any numerical digits in the query sequence should either be removed or replaced by appropriate letter codes (e.g., N for unknown nucleic acid residue or X for unknown amino acid residue).

Print out the description and sequences in the following format: 
```
- file_name | description | sequence_length | sequence
- file_name | description | sequence_length | sequence  
- file_name | description | sequence_length | sequence  
- file_name | description | sequence_length | sequence  
```

````python
# Solution here...

import re

def is_valid_sequence(sequence, sequence_type):
    '''
    Function for sequence format validation

    Inputs: 
      sequence (str): the sequence for analysis.
      sequence_type (str): nucleic or protein sequence type.
    
    Returns: 
      (bool), whether or not the sequence is valid
    '''

    # Use regular expressions from re package for matching
    # Documentation: https://docs.python.org/3/library/re.html#re.compile
    nucleic = re.compile(r"^[ACGTURYKMSWBDHVN\-]*$", re.IGNORECASE)
    protein = re.compile(r"^[ACDEFGHIKLMNPQRSTVWYUBZX\*-]*$", re.IGNORECASE)
    
    if sequence_type == "nucleic":
        return bool(nucleic.match(sequence))
    elif sequence_type == "protein":
        return bool(protein.match(sequence))
    else:
        raise ValueError(f"Sequence type '{sequence_type}' is invalid. Please specify 'nucleic' or 'protein'")

def fasta_parser(file_path, sequence_type):
    '''
    Fuction used as the FASTA file parser.

    Input: 
      file_path (str): path to your FASTA file
      sequence_type (str): nucleic or protein
        
    Returns: 
      (str): Parsed sequences (can be continuous with yield)
    '''
    
    description = None
    sequence = []

    # Read a FASTA formated file from disk and parse multiple outputs using 'yield' 
    #(help from https://www.w3schools.com/python/ref_keyword_yield.asp)
    with open(file_path, 'r') as fasta_file:
        for line in fasta_file:
            line = line.strip()
            if line.startswith(">"):
                # Yield the previous sequence if exists
                if description: 
                    yield (description, ''.join(sequence))
                # Start a new sequence if none exists
                description = line[1:].strip()
                sequence = []
            else:
                # Append sequence lines if not beggining line
                sequence.append(line)

        # Termination: Yield the last sequence once done reading
        if description:
            yield (description, ''.join(sequence))

def process_fasta_file(file_path, sequence_type):
    '''
    Main function to process and print FASTA files.

    Input: 
      file_path (str): path to your FASTA file
      sequence_type (str): nucleic or protein
        
    Returns: 
      errors out, or prints parsed sequence
    '''
    
    for description, sequence in fasta_parser(file_path, sequence_type):
        if not is_valid_sequence(sequence, sequence_type):
            print(f"Warning: Invalid sequence in {file_path}, description: {description}")
            return
        sequence_length = len(sequence)
        print(f"{file_path} | {description} | {sequence_length} | {sequence}")
````

Plot the distribution of the A,G,C,T bases for the SARS-CoV2 genome (sars-cov2-genome.fasta) using matplotlib.  Show the plot _inline_ in the notebook.

> Note that the exercise is to plot the data, not necessarily write a program. Use any tools at your disposal to acoomplisth this.

````python
%matplotlib inline

'''
My steps:
- Read the FASTA file and extract the sequence
- Count the occurrences of genome bases (A, G, C, T)
- Plot the base distribution

References: 
- Counter(): https://www.geeksforgeeks.org/python-count-occurrences-element-list/

'''

import matplotlib.pyplot as plt
from collections import Counter

def read_fasta_sequence(file_path):
    '''
    Function to read the FASTA file and extract the sequence.

    Input: 
      file_path (str): path to the fasta file
        
    Returns: 
      sequence (str): the sequence found
    '''
    
    sequence = []
    with open(file_path, 'r') as file:
        for line in file:
            # Skip the description line (which starts with ">")
            if line.startswith(">"):
                continue
            # Append sequence data (remove newlines and whitespaces)
            sequence.append(line.strip())
    return ''.join(sequence)

def count_bases(sequence):
    '''
    Function to count occurrences of each base

    Input: 
      sequence (str): the sequence found
        
    Returns: 
      (dict): the number of occurences
    '''
    return Counter(sequence.upper())

def plot_base_distribution(base_counts):
    '''
    Function to plot the base distribution using matplotlib

    Input: 
      base_counts (dict): the sequence found
        
    Returns: 
      nothing, just plots
    '''

    labels = base_counts.keys()
    values = base_counts.values()

    plt.bar(labels, values, color=['blue', 'orange', 'green', 'red'])
    plt.xlabel('Nucleotide Base')
    plt.ylabel('Count')
    plt.title('Distribution of bases A, G, C, T in SARS-CoV2 Genome')
    plt.show()

genome_sequence = read_fasta_sequence("sars-cov2-genome.fasta")
base_counts = count_bases(genome_sequence)
plot_base_distribution(base_counts)

````

Using your code from above, write a script that reads in a file in FASTA format, passes the description and sequence to your class, and determines the resulting protein sequence.
````python
# Class that emulates the central dogma of moleular biology
# Input: sequence and description 
# Functions: transcription from DNA to RNA and translation from RNA to protein. 

class CentralDogma:
    def __init__(self, description, sequence):
        self.description = description
        self.sequence = sequence.upper()

    def transcribe(self):
        '''
        Function to transcribe DNA to RNA (T -> U)
            
        Returns: 
          (str): sequence with replacements
        '''
        return self.sequence.replace('T', 'U')

    def translate(self, rna_sequence):
        '''
        Function to translate RNA to Protein
    
        Input: 
          self, rna_sequence (str): the RNA sequence to transcribe
            
        Returns: 
          (str): translated protein
        '''
        codon_table = {
            'AUG': 'M', 'UGG': 'W', 'UUU': 'F', 'UUC': 'F', 'UUA': 'L', 'UUG': 'L',
            'UCU': 'S', 'UCC': 'S', 'UCA': 'S', 'UCG': 'S', 'UAU': 'Y', 'UAC': 'Y',
            'UGU': 'C', 'UGC': 'C', 'UAA': '*', 'UAG': '*', 'UGA': '*', 'CUU': 'L',
            'CUC': 'L', 'CUA': 'L', 'CUG': 'L', 'CCU': 'P', 'CCC': 'P', 'CCA': 'P',
            'CCG': 'P', 'CAU': 'H', 'CAC': 'H', 'CAA': 'Q', 'CAG': 'Q', 'CGU': 'R',
            'CGC': 'R', 'CGA': 'R', 'CGG': 'R', 'AUU': 'I', 'AUC': 'I', 'AUA': 'I',
            'ACU': 'T', 'ACC': 'T', 'ACA': 'T', 'ACG': 'T', 'AAU': 'N', 'AAC': 'N',
            'AAA': 'K', 'AAG': 'K', 'AGU': 'S', 'AGC': 'S', 'AGA': 'R', 'AGG': 'R',
            'GUU': 'V', 'GUC': 'V', 'GUA': 'V', 'GUG': 'V', 'GCU': 'A', 'GCC': 'A',
            'GCA': 'A', 'GCG': 'A', 'GAU': 'D', 'GAC': 'D', 'GAA': 'E', 'GAG': 'E',
            'GGU': 'G', 'GGC': 'G', 'GGA': 'G', 'GGG': 'G'
        }
        protein = []
        for i in range(0, len(rna_sequence) - 2, 3):
            codon = rna_sequence[i:i + 3]
            if codon in codon_table:
                protein.append(codon_table[codon])
        return ''.join(protein)

    def find_orfs(self):
        '''
        Function identify open reading frames (ORFs)
           
        Returns: 
          (str): the open reading frames
        '''
        rna_sequence = self.transcribe()
        orfs = []
        for frame in range(3):
            orf = []
            start_pos = None
            for i in range(frame, len(rna_sequence) - 2, 3):
                codon = rna_sequence[i:i + 3]
                # Check for start codon
                if codon == 'AUG':
                    start_pos = i
                # Check for stop codon
                elif codon in ['UAA', 'UAG', 'UGA']:
                    if start_pos is not None:
                        orf = rna_sequence[start_pos:i + 3]
                        protein_sequence = self.translate(orf)
                        orfs.append((frame + 1, start_pos + 1, i + 3, len(protein_sequence), protein_sequence))
                    start_pos = None
        return orfs

def read_fasta_file(file_path):
    '''
    Function to read FASTA file and process it for its description and sequence

    Inputs:
      file_path (str): path to your fasta file
      
    Returns: 
      (str): the description and sequence
    '''
    description = None
    sequence = []
    with open(file_path, 'r') as file:
        for line in file:
            if line.startswith(">"):
                if description is None:
                    description = line[1:].strip()
            else:
                sequence.append(line.strip())
    return description, ''.join(sequence)

fasta_file = "../data/dogma.fasta.txt"
description, sequence = read_fasta_file(fasta_file)

# Instantiate CentralDogma class and find ORFs
central_dogma = CentralDogma(description, sequence)
orfs = central_dogma.find_orfs()

# Print ORFs in the specified format
for frame, start_pos, end_pos, protein_length, protein_sequence in orfs:
    print(f"* {frame} | {start_pos} | {end_pos} | {protein_length} | {protein_sequence}")
````
